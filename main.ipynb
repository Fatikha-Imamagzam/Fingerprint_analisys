```python
def select_image():
    dir_name = filedialog.askopenfilename(master=None, initialdir=os.getcwd())
    dir_name = os.path.dirname(dir_name)
    
    if dir_name.split("/"):
        dir_name = dir_name + "/"
    elif dir_name.split("\\"):
        dir_name = dir_name + "\\"
    
    return dir_name


def get_raceID_from_CSV(file, group, subgroup):
    df = pd.read_csv(file)
    df_subgroup = df.loc[df[group] == subgroup]
    id_subgroup = df_subgroup["id"]
    id_array = []
    for i in id_subgroup:
        id_array.append(i)
    return id_array


id_array = [get_raceID_from_CSV("Data_Files/participants.csv", "race", "white"),
            get_raceID_from_CSV("Data_Files/participants.csv", "race", "africanAmerican"),
            get_raceID_from_CSV("Data_Files/participants.csv", "race", "asian")]


def create_dir_save_img(dir_name, path, image):
    try:
        os.mkdir(dir_name) 
    except:
        pass
        
    img = Image.open(path)  
    img.save(dir_name + "/"+image)



    

def seperate_images():
    dir_name =  "Data_Files/B/roll/png/"#select_image()    
    c = 0
    print(f"{end[5]}Getting Images Ready For Seperation...")
    for file in os.listdir(dir_name):
        if file.endswith(".jpg") or file.endswith(".jpeg") or file.endswith(".png"):
            print(os.path.splitext(file)[0])
            im_name = os.path.splitext(file)[0]
            print(dir_name," ",im_name)
            path = dir_name + os.path.splitext(file)[0] + os.path.splitext(file)[1]
            image  =  os.path.splitext(file)[0] + os.path.splitext(file)[1]
            print(image," ",path)
            print(im_name[0:4])
            currentID = str(im_name[4:8])
            print(currentID) 
            
            for i in id_array[0]:
                if str(i) == currentID:
                    id_w = currentID
                    print("w: ", id_w)  
                    dest_dir = "Data_Files/race/white"
                    create_dir_save_img(dest_dir, dir_name+image, image)
                    
                else:
                    pass
            for i in id_array[1]:
                if str(i) == currentID:
                    id_af_a = currentID
                    print("af_a: ", id_af_a)
                    dest_dir = "Data_Files/race/african american" 
                    create_dir_save_img(dest_dir, dir_name+image, image)                   
                else:
                    pass
            for i in id_array[2]:
                if str(i) == currentID:
                    id_a = currentID
                    print("a: ", id_a)
                    dest_dir = "Data_Files/race//asian"
                    create_dir_save_img(dest_dir, dir_name +image, image)                    
                else:
                    pass
            
             
            
seperate_images()
```

```python
import os
import cv2
import PIL
import glob
import pathlib
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import imgaug.augmenters as iaa
from tensorflow.keras.optimizers import RMSprop
from tensorflow.keras.preprocessing import image
from sklearn.metrics import classification_report, accuracy_score 
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from datetime import datetime 
end = ["\033[0;30;1m","\033[1;31;1m","\033[1;32;1m", "\033[1;33;1m", "\033[1;34;1m", "\033[1;35;1m"] 

def train_model():
    start_time = datetime.now()
    print(f"{end[2]}Traiining starts soon...........{start_time}\n\n")
     
    train = ImageDataGenerator(rescale=1/255)
    validation = ImageDataGenerator(rescale=1/255)


    train_dataset = train.flow_from_directory("./Data_files/gender/training/",
                                             target_size = (512, 512),
                                             batch_size = 32,
                                             class_mode = "binary")

    validation_dataset = validation.flow_from_directory("./Data_files/gender/validation/",
                                             target_size = (512, 512),
                                             batch_size = 32,
                                             class_mode = "binary")


    try:
        checkpoint_ = os.mkdir(os.getcwd() +"\checkpoint6\\")
    except FileExistsError as FER:
        checkpoint_ = os.getcwd() +"\checkpoint6\\"
        print("'./checkpoint6' Already Exist!")

    cwd = os.getcwd()
    print(cwd+"\checkpoint6")
    checkpoint_path =  str(checkpoint_) 
    checkpoint_dir = os.path.dirname(checkpoint_path)
    print(checkpoint_dir)
    # Create a callback that saves the model's weights
    cp_callback = tf.keras.callbacks.ModelCheckpoint(filepath=checkpoint_dir,
                                                     save_weights_only=True,
                                                     verbose=1)



    # Create a callback that saves the model's weights
    cp_callback = tf.keras.callbacks.ModelCheckpoint(filepath=checkpoint_,
                                                     save_weights_only=True,
                                                     verbose=1)
   # model = create_model() 
    
    model = tf.keras.models.Sequential([tf.keras.layers.Conv2D(16, (3,3), activation = "relu", input_shape=(512,512, 3)),
                                    tf.keras.layers.MaxPool2D(2,2),
                                    #
                                    tf.keras.layers.Conv2D(32, (3,3), activation = "relu"),
                                    tf.keras.layers.MaxPool2D(2,2),
                                    #
                                    tf.keras.layers.Conv2D(64, (3,3), activation = "relu"),
                                    tf.keras.layers.MaxPool2D(2,2),
                                    ##
                                    tf.keras.layers.Flatten(),
                                    ##
                                    tf.keras.layers.Dense(512, activation = "relu"),
                                     ##
                                    tf.keras.layers.Dense(1, activation = "sigmoid")
                                   ])
    
    model.compile(loss="binary_crossentropy",
             optimizer = RMSprop(learning_rate=0.1),
              metrics = ["accuracy"]
             )
    model.summary()
    model_filt = model.fit(train_dataset, 
                          epochs = 10,
                           callbacks=[cp_callback],
                           validation_data = validation_dataset)
    
    time_elapsed = datetime.now() - start_time
    print('\n\nTraining Time (hh:mm:ss.ms) {}'.format(time_elapsed))
    return model


if __name__ =="__main__":
    selection = input(f"What Do You Want To Do?:\n{end[1]}Train A Model(T)\n{end[4]}Load Pre-Trained Model(L)\n>>")
    if selection.upper() == "L":
        model = create_model() 
        model.load_weights("./checkpoint/")
    elif selection.upper() == "T":
        confirm = input("Are You Sure You WantTo Train A Model? Yes(Y)/No(N)")
        if confirm.upper()=="Y":
            train_model()
        else:
            pass
```
